```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = np.arange(1, 30)
f = (N+1)/2
g = N/2

plt.plot(f)
plt.plot(2*g)
plt.legend(["$f$", "$g$"])
```

### $10*log2(N)$ is $O(N)$

```python
N = np.arange(1, 200)
f = 10*np.log2(N)
g = N

plt.plot(f)
plt.plot(g)
plt.plot([80, 80], [0, 200])
plt.legend(["$f = 10 \\log_2(N)$", "$g = N$"])
```
